```python
folder_path = 'testcsv'

csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

df_list = [pd.read_csv(os.path.join(folder_path, file)) for file in csv_files]
merged_movie_list = pd.concat(df_list, ignore_index=True)

merged_movie_list.to_csv('merged_movie_file.csv', index=False)
```

```python
import pandas as pd
import re

df = pd.read_csv('merged_movie_file.csv')

df['Title'] = df['Title'].str.replace(r'^\d+\.\s*', '', regex=True)

df.to_csv('merged&cleaned_movies.csv', index=False)
```

```python
import pandas as pd

df = pd.read_csv('merged&cleaned_movies.csv')

df['Voting'] = df['Voting'].astype(str)

df['Voting'] = df['Voting'].str.replace(r'[\(\)]', '', regex=True)

df['Voting'] = df['Voting'].str.replace(r'^-', '', regex=True)

df.to_csv('merged&cleaned_movies1.csv', index=False)
```

```python
import pandas as pd


df = pd.read_csv('merged&cleaned_movies1.csv')

def convert_k_to_int(value):
    value = str(value).strip().upper()
    if 'K' in value:
        try:
            return int(float(value.replace('K', '')) * 1000)
        except ValueError:
            return None
    else:
        try:
            return int(value)
        except ValueError:
            return None

df['Voting'] = df['Voting'].apply(convert_k_to_int)

df.to_csv('Cleaned_movies_list.csv', index=False)
```
